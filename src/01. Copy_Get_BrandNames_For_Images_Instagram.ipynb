```python
# data cleaning beverage
data5['beverage_full_brand_names'].replace(" Bigelow's Tea", "Bigelow's Tea", inplace=True) #bigelow's aposthrophe different to same 
data7['brand_full_name'].replace(' Bigelow`s Tea', "Bigelow's Tea", inplace=True) #bigelow's secret's aposthrophe different to same 
data8['brand_full_name'].replace(' Bigelow`s Tea', "Bigelow's Tea", inplace=True) #bigelow's secret's aposthrophe different to same 

data5['beverage_full_brand_names'].replace(" Jack Daniel's", "Jack Daniel's", inplace=True) #jack daniel's aposthrophe different to same 
data7['brand_full_name'].replace(' Jack Daniel`s', "Jack Daniel's", inplace=True) #jack daniel's aposthrophe different to same
data8['brand_full_name'].replace(' Jack Daniel`s', "Jack Daniel's", inplace=True) #jack daniel's aposthrophe different to same

data5['beverage_full_brand_names'].replace(" Mott's", "Mott's", inplace=True) #mott's aposthrophe different to same 
data7['brand_full_name'].replace(' Mott`s', "Mott's", inplace=True) #mott's aposthrophe different to same 
data8['brand_full_name'].replace(' Mott`s', "Mott's", inplace=True) #mott's aposthrophe different to same 

data5['beverage_full_brand_names'].replace(" Welch's", "Welch's", inplace=True) #welch's aposthrophe different to same 
data7['brand_full_name'].replace(' Welch`s', "Welch's", inplace=True) #welch's aposthrophe different to same 
data8['brand_full_name'].replace(' Welch`s', "Welch's", inplace=True) #welch's aposthrophe different to same

data5['beverage_full_brand_names'].replace(' Monster', 'Monster', inplace=True) #monster with no whitespace 
data7['brand_full_name'].replace(' Monster (drink)', 'Monster', inplace=True) #remove (drink) from Monster
data8['brand_full_name'].replace(' Monster (drink)', 'Monster', inplace=True) #remove (drink) from Monster

data5['beverage_full_brand_names'].replace(' Vitaminwater', 'Vitaminwater', inplace=True) #vitaminwater with no whitespace 
data7['brand_full_name'].replace(' vitaminwater', 'Vitaminwater', inplace=True) #vitaminwater lowercase v to uppercase V
data8['brand_full_name'].replace(' vitaminwater', 'Vitaminwater', inplace=True) #vitaminwater lowercase v  to uppercase V
```

Put all firm account names (of all brand categories) into a list, which can be iterated over to get images from these firm accounts on Instagram. It is used to specify user-profiles Instagram pages. 

```python
firm_usernames = data7.firm_account.values.tolist()
```

```python
def write_list_to_file(guest_list, filename):
    """Write the list to csv file."""

    with open(filename, "w") as outfile:
        for entries in guest_list:
            outfile.write(entries)
            outfile.write("\n")

write_list_to_file(firm_usernames, 'firm_usernames.csv')      
```
